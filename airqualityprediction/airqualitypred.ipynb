```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X=pd.read_csv('./Train/Train.csv').values
Y=pd.read_csv('./Test/Test.csv').values
```

```python
print(X.shape)
```

```python
x=X[:,0:5]
```

```python
print(x.shape)
```

```python
y=X[:,5]
```

```python
theta=np.dot(np.linalg.pinv(np.dot(x.T,x)),np.dot(x.T,y))
```

```python
theta
```

```python
y_ = np.dot(Y,theta)
```

```python
df=pd.DataFrame(y_)
```
